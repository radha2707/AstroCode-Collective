```python
import numpy as np
import matplotlib.pyplot as plt

# Load the first .dat file
file1_path = '/ca24/AstroCode-Collective/HD_209458_b_spectrum_assignment3_taskB.dat'  # Replace with your first file's path
data1 = np.loadtxt(file1_path)

# Load the second .dat file
file2_path = '/ca24/AstroCode-Collective/Simple_Transmission_retrieval_TaskD_updated.dat'  # Replace with your second file's path
data2 = np.loadtxt(file2_path)

# Extract columns for the first file
wavelength1 = data1[:, 0]  # First column: Wavelength
transit_depth1 = data1[:, 1]  # Second column: Transit depth
errors1 = data1[:, 2]  # Third column: Error

# Extract columns for the second file
wavelength2 = data2[:, 0]  # First column: Wavelength
transit_depth2 = data2[:, 1]  # Second column: Transit depth
errors2 = data2[:, 2]  # Third column: Error

# Plot both datasets
plt.figure(figsize=(10, 6))

# Plot first dataset
plt.errorbar(wavelength1, transit_depth1, yerr=errors1, fmt='o', capsize=3, label='Original Spectrum', color='blue')
#plt.plot(wavelength1, transit_depth1)
# Plot second dataset
plt.errorbar(wavelength2, transit_depth2, yerr=errors2, fmt='s', capsize=3, label='Retrieved Spectrum', color='red')
#plt.plot(wavelength2, transit_depth2)
# Add labels, title, legend, and grid
plt.xlabel('Wavelength')
plt.ylabel('Transit Depth')
plt.title('Comparison of Transit Depth vs. Wavelength')
plt.grid(True)
plt.legend()
plt.show()
```
